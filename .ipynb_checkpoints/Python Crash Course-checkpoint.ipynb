```python
# Lists & Arrays

# We use square brackets to tell python when we're making a list or array.
my_list = [4,5,7,'Annapoorani'] # <--------- Lists can contain many pieces of data, with varied data types if necessary. We'll show how to get stuff 
# out of them in 'indexing' cell. We call each piece of data in a list an 'element.'

# To append to the list, we use:
my_list.append('hi') # <------- this will go after 'Annapoorani' (and will get deleted in the next line)

# To remove from the list, we use indexing, so come back once you've reached that cell:
my_list.pop(4) # <---------- No square brackets

# We could do this, too: 
# To get the length of the list:
print("The length of my list is",len(my_list))

my_array = [[4,5,7,'Annapoorani'],[3,2,'hello world',17]] # <------------- An array is a list of lists. This one has 2 lists, each with 4 elements, so it
# is a 2 by 4 array (which is 2D).
three_d_array = [
    [[5,8],[7,3]],
    [[2,6],[4,5]],
    [[9,8],[7,1]]
] # <----------- This is a 3 (3 main lists) by 2 (2 lists in each of the three main lists) by 2 (2 elements in each inner list) array. It's 3d, 
# since there are 3 opening brackets, and three closing brackets:
three_d_array
```

```python
# Indexing
# Python starts counting from 0. So, if we want the first element in our list, we're actually looking for element #0.

# We get elements like this, using square brackets after the list name:
print("The first element of my list is",my_list[0])

# To index out of an array, we first grab one of the inner lists, then index normally off of that inner list, like so:
print("The first element in the first list in my array is",my_array[0][0])

# Or, for the three dimensional array:
print("The first element in the first inner list in the first outer list in my 3D array is",three_d_array[0][0][0])
print("\n")

# We can also grab slices of a list, like so:
print("The first three elements in my list are",my_list[:3]) # This will stop right before showing element #3 - which is actually the 4th element.
print("The elements from #1 (2) to #2 (3) in my list are",my_list[1:3]) # This will start from element #1 and stop before element #3 - the fourth element.
print("\n")
print("This is all the elements but the last one:",my_list[:-1]) # Basically, this goes until len(list)-1
print("This is the last element:",my_list[-1:]) # And this starts from len(list)-1
```

```python
# For loops

for i in range(3):
    print("i is",i)
print("\n")
# The range is the # of times the list will run. i starts from 0. 

for element in my_list:
    print("The current element is",element)
# We don't have to use ranges. We can tell it to go through a list if we want to. Now that we don't have to use 'element' - that's
# just a variable that takes on different values.

print("\n")
for number in my_list[:-1]:
    print("The current number is",number)
```

```python
# Imports
import math # <--------- This imports a set of commands that Python can now use - such as square root.
print(math.sqrt(16))  # We use 'library.' whenever we call a command from a given library

import random # <----------- This helps us get random numbers.
print(random.randint(1, 10))
```

```python
# Dictionaries

# Dictionaries are a special data type that will pop up occasionally in this course.
me = {
    'my name': 'Annapoorani',
    'my_age': 15,
    'My home state': 'Washington',
    'Some of my hobbies': ['Eating','Reading','Making Udemy Courses']
}
# Each 'key', or attribute name, is followed by a colon, after which comes its value. The value can be anything, from integers to lists.

# You can access the info using list indexing, but with the key name instead of index number.
print(me['my name'])

# You can add info by just creating a new key:
me['personality type'] = 'ENTJ'
me
```

```python
# Conditionals
my_name = 'Annapoorani'
my_age = 15
if my_name == 'Annapoorani': # We use == to check if a variable is equal to something, since = would reassign it.
    print('Hi Annapoorani')
elif my_name == 'Franklin' and my_age == 15: # We can use elif if we want to check more than just an if and else condition. Additionally, we use 
# and if we want to check if two things are simultaneously true.
    print('Hi 15 year old Franklin') 
else:
    print("I don't know your name, but hi!")
```

```python
# Functions

def print_my_info(age, place_of_birth, name, favorite_food="Don't have one"): # <----- This is a function. It's a chunk of code we may want to 
# re-use several times, so we save in a function.
# Notice there are variable names inside the parentheses. These can change every time we call the function - we just have to specify them, in order. 
# These are called arguments, and specifying them is referred to as 'passing in an argument'. The argument or arguments at the end can have default 
# values - values that will apply if you don't specify something else.
    print("Your age is:",age) # <----- We can use the variables in the parentheses in our functions.
    print("Your place of birth is:",place_of_birth)
    print("Your name is:",name)
    print("Your favorite food is:",favorite_food) # This will default to "Don't have one" if the value is not specified.
print_my_info(15, 'Washingon','Annapoorani','pasta')
print("")
print_my_info(13,'California','Sally') # This will default to "Don't have one" since the value is not specified and the favorite_food argument happens to have a default.
# YOU MUST PASS ARGUMENTS IN THE RIGHT ORDER!
```
