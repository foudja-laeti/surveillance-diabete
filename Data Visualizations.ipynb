Code Lecture 2 - Histograms

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['Age'],bins=10,edgecolor='black')
plt.show()
```

```python
sns.histplot(data=df,x='Age',bins=10)
plt.show()
```

```python
sns.kdeplot(data=df,x='Age')
plt.show()
```

```python
sns.kdeplot(data=df[df['Outcome']==1],x='Glucose',label='Diabetic')
sns.kdeplot(data=df[df['Outcome']==0],x='Glucose',label='Non-diabetic')
plt.title('Glucose Level Distribution by Outcome')
plt.legend()
plt.show()
```

Code Lecture 3 - Scatterplots & Subplots

```python
plt.scatter(df['Glucose'],df['BMI'])
plt.title('Glucose vs. BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

```python
sns.scatterplot(data=df,x='Glucose',y='BMI',hue='Outcome')
plt.title('Glucose vs. BMI')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data=df,x='Glucose',y='BMI',hue='Outcome',ax=ax)
ax.set_title('Glucose vs. BMI by Outcome')
st.pyplot(fig)
fig
```

Code Lecture 5 - Data Cleaning

```python
df.head()
```

```python
df_temp = df.drop('Outcome',axis=1)
```

```python
df_temp = df_temp.replace(0,np.nan)
```

```python
sns.heatmap(df_temp.isnull(),cmap='viridis')
plt.show()
```

```python
df_temp = df.drop(columns=['Outcome','Pregnancies','Insulin','SkinThickness'])
df_temp = df_temp.replace(0,np.nan)
df = pd.concat([df['Pregnancies'],df['Insulin'],df['SkinThickness'],df_temp,df['Outcome']],axis=1)
```

```python
df = df.dropna().reset_index(drop=True)
```
